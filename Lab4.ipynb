```python
def createMatrix(n):
    m=[]
    for i in range (n):
        c=[]
        for j in range(n):
            c.append(0)
        m.append(c)
    return m

def printMatrix(m):
    for i in range(len(m)):
        print(m[i])

def inDegree(m,v):
    c=0
    for i in range(len(m[v])):
        if m[i][v]==1:
            c=c+1
    return c
            

def outDegree(m,v):
    c=0
    for i in range(len(m[v])):
        if m[v][i]== 1:
            c=c+1
    return c

def maxWeight(mat):
    z=0
    row=0
    col=0
    for i in range(len(mat)):
        for j in range(len(mat)):
            if mat[i][j]>z:
                z=mat[i][j]
                row=i
                col=j
    return row,col,z


def minWeight(mat):
    z=100
    row=0
    col=0
    for i in range(len(mat)):
        for j in range(len(mat)):
            if mat[i][j]<z and mat[i][j]>0:
                z=mat[i][j]
                row=i
                col=j
    return row,col,z


text = open("graph3.txt")

line= text.readline()
l=line.split()

n=int(l[0])
m=int(l[1])

matrix = createMatrix(n)
printMatrix(matrix)
print(" ")


for i in range(m):
    line= text.readline()
    u = int(line.split()[0])
    v = int(line.split()[1])
    w = int(line.split()[2])
    
    matrix[u][v] = w
    
printMatrix(matrix)

print(outDegree(matrix,5))
print(inDegree(matrix,5))
print(maxWeight(matrix))
print(minWeight(matrix))
```
