```python
import queue

def createList(n):
    g = {}
    for i in range(n):
        g[i] = []  
    return g

def printList(g):
    for node in g:
        print(f"{node} -> {g[node]}")

def ucs(G, s, n, goal):
    q = queue.PriorityQueue()  
    visited = [False] * n 
    q.put((0, s, []))

    while not q.empty():
        gn, node, path = q.get()  

        if visited[node]:
            continue  

        visited[node] = True  # Mark as visited
        path = path + [node]  # Update the path

        # If the goal is reached, return the cost and path
        if node == goal:
            print("Path:", path)
            return gn

        print(f"Exploring Node: {node}, Cost: {gn}")

        # Explore all neighbors
        for neighbor, weight in G[node]:
            if not visited[neighbor]:
                q.put((gn + weight, neighbor, path))  # Push with cumulative cost

    print("Goal not reachable.")
    return -1

# Read graph data from file
f = open("graph2.txt")
line = f.readline().split()
n = int(line[0])  # Number of nodes
m = int(line[1])  # Number of edges

adjList = createList(n)

# Populate the adjacency list with weights
for i in range(m):
    line = f.readline().split()
    u = int(line[0])  # Start node
    v = int(line[1])  # End node
    w = int(line[2])  # Weight
    adjList[u].append((v, w))  # Add edge as a tuple (neighbor, weight)

printList(adjList)

ucs(adjList, 0, n, 4)
```
